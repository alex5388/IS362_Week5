#What is the northernmost airport in the United States? EEN

To answer questions What is the northernmost airport in the United States? 
I think we can sort from ascending to descending latitude to get the northernmost airport. 

```python
airports.sort_values(by='lat', ascending=False)
```

A quick google search confirms that Dillant Hopkins Airport is located in New Hampshire. Since he came up as the highest latitude, then it proves that it's the most northern airport on our list.

#What is the easternmost airport in the United States? EPM

2. What is the easternmost airport in the United States?
Longitude 0 is in England so in order to answer our question we will actually be looking for a western longitude. The eastern most part of the US is West Quoddy Head, Maine at -66.95, so we will be looking for the closest value to -66.

```python
airports[(airports.lon < -65)].sort_values(by='lon', ascending=False)
```

I sorted the airports by those with a longitude that is less than -65. -65 is east of -66, so eliminating everything east of -65 would leave no room for any surprise values in -66 that might fall outside of our range. The easternmost airport is Eastport Municipal Airport.

#On February 12th, 2013, which New York area airport had the windiest weather? EWR

3. On February 12th, 2013, which New York area airport had the windiest weather?
To find this answer we have to pull out all data on February 12, 2013 for New York airports. Then we can sort by highest wind speed.

```python
feb12 = weather[(weather.year == 2013) & (weather.month == 2) & (weather.day == 12)]
```

```python
feb12.sort_values(by='wind_speed', ascending=False)
```

After pulling only 2/12/2013 weather data and then sorting it by wind speed, we can see that EWR has the highest wind speed on that day.
